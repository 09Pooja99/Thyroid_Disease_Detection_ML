```python
import pandas as pd
import os
import zipfile
import numpy as np
import requests
```

```python


# URL of the dataset
url = "https://archive.ics.uci.edu/static/public/102/thyroid+disease.zip"

# Download the ZIP file
response = requests.get(url)
zip_file_path = "thyroid_disease.zip"

# Save the file locally
with open(zip_file_path, "wb") as file:
    file.write(response.content)

print("✅ ZIP file downloaded successfully!")
```

```python


# Extract all files to a directory
extract_dir = "thyroid_data"
with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
    zip_ref.extractall(extract_dir)

# List all extracted files
extracted_files = os.listdir(extract_dir)
print("Extracted Files:", extracted_files)
```

```python
# Define file path
data_file = os.path.join(extract_dir, "allbp.data")

columns = [ 
    'age', 'sex', 'on thyroxine', 'query on thyroxine', 'on antithyroid medication',
    'sick', 'pregnant', 'thyroid surgery', 'I131 treatment', 'query hypothyroid',
    'query hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych',
    'TSH measured', 'TSH', 'T3 measured', 'T3', 'TT4 measured', 'TT4', 'T4U measured',
    'T4U', 'FTI measured', 'FTI', 'TBG measured', 'TBG', 'referral source', 'Class'
]

# Load dataset
df = pd.read_csv(data_file, delimiter=",", header=None, na_values="?",names=columns)
csv_file_path = "allbp.csv"
df.to_csv(csv_file_path, index=False)

# Show first 5 rows
#print(df.head())
print(df.shape)
```
